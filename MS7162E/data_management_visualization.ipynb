# October 24 2024 - scatter plot using R
Scatter plot plots the coordinates of the two features of an observation and both the features are ideally continuous observations

```r
# load data
pdat  <- base::readRDS(url("https://slcladal.github.io/data/pvd.rda", "rb"))
```

```r
library(ggplot2)
```

```r
#Aesthetic mappings (aes)describe how variables in the data are mapped to visual properties (aesthetics) of geoms
ggplot(pdat, aes(x = Date, y = Prepositions))
```

```r
# The point geom (geom_point()) is used to create scatterplots
ggplot(pdat, aes(x = Date, y = Prepositions)) +
  geom_point()
```

```r
#Smoothed conditional means is given by geom_smooth
ggplot(pdat, aes(x = Date, y = Prepositions)) +
  geom_point() +
  geom_smooth(se = F)
```

```r
#The classic dark-on-light ggplot2 theme theme_bw()
ggplot(pdat, aes(x = Date, y = Prepositions)) +
  geom_point() +
  geom_smooth(se = F) +
  theme_bw()
```

```r
# store plot in object p
p <- ggplot(pdat, aes(x = Date, y = Prepositions)) +
  geom_point() +
  theme_bw()
# add layer with nicer axes titles to p
p + labs(x = "Year", y = "Frequency")


```

```r
p + labs(x = "Year", y = "Frequency") +
ggtitle("Preposition use over time", subtitle="based on the PPC corpus")
```

# October 29 2024 - bar plot using R

Bar plot
Coded by Arjun Anil Kumar <br>
Bar plots shows the variation of a count variable with respect to a discrete ordinal or nominal variable

```r
# Create the data for the chart
sales_data <- c(17, 32, 8, 53, 1)
```

```r
# Plot the bar chart
barplot(sales_data,
        xlab = "Years",
        ylab = "Sales Revenue",
        main ="Years and Sales Revenue",
        col = "green")
```

```r
# declare the variable years
years <- c("1999","2000","2001","2002","2003")
class(years)
# convert years to factor variable
years <- as.factor(years)
class(years)
barplot(sales_data,
        xlab = "Years",
        ylab = "Sales Revenue",
        main ="Years and Sales Revenue",
        col = "green")
# label the xaxis with years
axis(side = 1, at = c(1,2,3,4,5), labels = years )
```

```r
# bar plot using ggplot
# selectively color the bar plots based on the condition
library(ggplot2)
sales_data_frame <- data.frame(years,sales_data)
threshold_sales <- 10
# geom_col is for bar plots where the y coordinate represents the count stored in the data
ggplot(sales_data_frame, aes(x = years, y =sales_data, fill = sales_data)) +
  geom_col()
```

```r
# color using hexcode
ggplot(sales_data_frame, aes(x = years, y =sales_data, fill = sales_data)) +
  geom_col(fill="#FFCC00")
```

```r
# selective coloring based on the threshold sales
ggplot(sales_data_frame, aes(x = years, y =sales_data, fill = sales_data)) +
  geom_col()+
geom_col(fill=ifelse(sales_data_frame$sales_data < threshold_sales, 'gray', NA))
```

```r
# Add title and x and y labels
ggplot(sales_data_frame, aes(x = years, y =sales_data, fill = sales_data)) +
  geom_col()+
  geom_col(fill=ifelse(sales_data_frame$sales_data < threshold_sales, 'gray', NA))+
  ggtitle("Bar plot Sales 1999-2003")+
  xlab('Years')+
  ylab('Sales Data')
```

```r
# Center the title of the bar plot
ggplot(sales_data_frame, aes(x = years, y =sales_data, fill = sales_data)) +
  geom_col()+
  geom_col(fill=ifelse(sales_data_frame$sales_data < threshold_sales, 'gray', NA))+
  ggtitle("Bar plot Sales 1999-2003")+
  theme(plot.title = element_text(hjust=0.5))
```

```r
# bar plot customization
# bar plots practice
# Data with category and count
df1 <- data.frame(group = c("A", "B", "C"),
                 count = c(3, 5, 6))
library(ggplot2)
# use stat = identity if the count has already been computed
ggplot(df1, aes(x = group, y = count)) +
  geom_bar(stat = "identity")
# geom_col is the same as geom_bar(stat = "identity")
ggplot(df1, aes(x = group, y = count)) +
  geom_col()
# horizontal bar plot - coord_flip
library(ggplot2)
ggplot(df1, aes(x = group, y = count)) +
  geom_bar(stat = "identity") +
  coord_flip()+
  ggtitle('horizontal bar')
# orders of the bar in the bar plot using scale_x_discrete
ggplot(df1, aes(x = group, y = count)) +
  geom_bar(stat = "identity") +
  scale_x_discrete(limits = c("C", "B", "A"))+
  ggtitle('customized order of cateogorical variables')
# adding labels to the bar
ggplot(df1, aes(x = group, y = count)) +
  geom_bar(stat = "identity") +
  geom_text(aes(label = count), vjust = 2, colour = "white")+
  ggtitle('adding labels to bar')
# coloring the bar with blue color
ggplot(df1, aes(x = group, y = count)) +
  geom_bar(stat = "identity", fill = 4)+
  ggtitle('color bar with default color')
# color by group
ggplot(df1, aes(x = group, y = count, fill = group)) +
  geom_bar(stat = "identity") +
  ggtitle('color by group')
# specify a color for a group
ggplot(df1, aes(x = group, y = count, fill = group)) +
  geom_bar(stat = "identity") +
  scale_fill_manual(values = c("#FCFED4", "#CCEDB1", "#41B7C4"))+
  ggtitle('specified color for a group')
# legend customization
ggplot(df1, aes(x = group, y = count, fill = group)) +
  geom_bar(stat = "identity") +
  theme(legend.position = "bottom")+
  ggtitle('Legend customization')
# legend title
ggplot(df1, aes(x = group, y = count, fill = group)) +
  geom_bar(stat = "identity") +
  guides(fill = guide_legend(title = "Group Legend")) +
  ggtitle('Legend title changed')

```
